```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import numpy as np
```

```python
def plotaArrumado(train_scores, test_scores, num_epochs):
    import numpy as np
    i =0
    for ar in train_scores:
        train_scores[i] = ar[:num_epochs]
        i+=1
    i =0
    for ar in test_scores:
        test_scores[i] = ar[:num_epochs]
        i+=1
    import matplotlib.pyplot as plt
    # Set up the plot
    plt.figure()
    plt.title('K-fold losses')
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.grid()

    import numpy as np
    
    # Calculate mean and distribution of training history
    train_scores_mean = np.mean(train_scores, axis=0)
    train_scores_std = np.std(train_scores, axis=0)
    test_scores_mean = np.mean(test_scores, axis=0)
    test_scores_std = np.std(test_scores, axis=0)

    # Plot the average scores
    plt.plot(
        train_scores_mean,
        '-',
        color="b",
        label="Training loss"
    )
    plt.plot(
        test_scores_mean,
        '-',
        color="g",
        label="Validation loss"
    )

    # Plot a shaded area to represent the score distribution
    epochs = list(range(num_epochs))
    plt.fill_between(
        epochs,
        train_scores_mean - train_scores_std,
        train_scores_mean + train_scores_std,
        alpha=0.1,
        color="b"
    )
    plt.fill_between(
        epochs,
        test_scores_mean - test_scores_std,
        test_scores_mean + test_scores_std,
        alpha=0.1,
        color="g"
    )

    plt.legend(loc="lower right")
    plt.show()
```

```python

for filename in os.listdir('logy'):
    print(filename)
    min_epochs = 100
    train_scores = []
    test_scores = []
    for cont in range(5):
        print(cont)
        if os.path.exists('logy/'+filename+'/'+str(cont)+'/Loss'):
            event_loss = EventAccumulator('logy/'+filename+'/'+str(cont)+'/Loss/val_loss')
            event_loss.Reload()
            event_train_loss = EventAccumulator('logy/'+filename+'/'+str(cont)+'/Loss/train_loss')
            event_train_loss.Reload()
            w_times, step_nums, valloss = zip(*event_loss.Scalars('Loss'))
            w_times, step_nums, trainloss = zip(*event_train_loss.Scalars('Loss'))
            if len(trainloss) < min_epochs:
                min_epochs = len(trainloss)
            train_scores.append(np.array(trainloss))
            test_scores.append(np.array(valloss))
            # plt.plot(valloss)
            # plt.plot(trainloss)
            # # plt.plot(valsacc)
            # # plt.plot(trainacc)
            # plt.title(filename + ' Loss')
            # plt.ylabel('Loss')
            # plt.xlabel('epoch')
            # plt.legend(['val_loss', 'train_loss','val_acc', 'train_acc'], loc='upper left')
            # plt.savefig('logy/'+filename+'/history_'+filename+'_fold'+str(cont)+'.png')
            # plt.show()

    plotaArrumado(train_scores, test_scores, min_epochs)
```
